```python
import pandas as pd
import pandas_datareader as pdr
import datetime as dt
import numpy as np
```

```python
tickers = ['AAPL', 'TWTR', 'IBM', 'MSFT']
start = dt.datetime(2020, 1, 1)

data = pdr.get_data_yahoo(tickers, start)
```

```python
data.head()
```

```python
data = data['Adj Close']
```

```python
data.head()
```

```python
log_returns = np.log(data/data.shift())
```

```python
log_returns
```

```python
log_returns.corr()
```

```python
sp500 = pdr.get_data_yahoo("^GSPC", start) #The S&P 500 Index features 500 leading U.S. publicly traded companies
```

```python
sp500.head()
```

```python
log_returns['SP500'] = np.log(sp500['Adj Close'] / sp500['Adj Close'].shift())
```

```python
log_returns.corr()
```

```python
def test_correlation(ticker):
    df = pdr.get_data_yahoo(ticker, start)
    lr = log_returns.copy() #entire market data
    lr[ticker] = np.log(df['Adj Close'] / df['Adj Close'].shift()) #entering input market data
    return lr.corr() #returning all of their correlation
```

```python
test_correlation("LQD")
```

```python
test_correlation("TLT")
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def visualize_correlation(ticker1, ticker2):
    df = pdr.get_data_yahoo([ticker1, ticker2], start)
    df = df['Adj Close']
    df = df/df.iloc[0]
    fig, ax = plt.subplots()
    ax.set_labely("correlation between {} and {}".format(ticker1, ticker2))
    df.plot(ax=ax)
```

```python
visualize_correlation('AAPL', 'TLT')
```

```python
visualize_correlation("^GSPC", "AAPL")
```
